```python
### global variabel
Amount = 10000
portfolio = 0
total_portfolio_val = 0
money_end = Amount
investment = []
transection_cost = 0.0075
```

```python
def add_money(money:int):
    global Amount,money_end
    Amount +=money
    money_end +=money
    return 'Money added...'
```

```python
def  buy(quantity, price):
    global portfolio,money_end
    allocated_money = quantity*price
    allocated_total_money = allocated_money + transection_cost*allocated_money

    if allocated_total_money>money_end:
        return 'Transection not completed...\nInsuficient amount.'
    
    money_end = money_end - allocated_money - transection_cost*allocated_money
    portfolio += quantity

    if investment ==[]:
        investment.append(allocated_money)
    else:
        investment.append(allocated_money)
        investment[-1] += investment[-2]
```

```python
def sell(quantity, price):
    global portfolio, money_end
    allocated_money = quantity*price
    money_end = money_end + allocated_money - transection_cost* allocated_money
    portfolio -=quantity
    investment.append(-allocated_money)
    investment[-1] += investment[-2]
```

```python
price = 100   ##Price of the stock
quantity = 4   ### quantity of the the stockes
buy(quantity,price)
```
